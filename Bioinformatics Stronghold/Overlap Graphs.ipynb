```python
#####################
# code to read FASTA file
#####################
import sys, re

f=open('rosalind_grph(4).txt','r')
lines=f.readlines()

hre=re.compile('>(\S+)')
lre=re.compile('^(\S+)$')

gene={}

for line in lines:
        outh = hre.search(line)
        if outh:
                id=outh.group(1)
        else:
                outl=lre.search(line)
                if(id in gene.keys()):
                        gene[id] += outl.group(1)
                else:
                        gene[id]  =outl.group(1)
                        
seq = list(gene.values())
id = list(gene.keys())

for i in range(len(seq)):
    for j in range(len(seq)):
        if (i != j) and (seq[i][-3:] == seq[j][:3]):
            print(id[i], id[j])
```
